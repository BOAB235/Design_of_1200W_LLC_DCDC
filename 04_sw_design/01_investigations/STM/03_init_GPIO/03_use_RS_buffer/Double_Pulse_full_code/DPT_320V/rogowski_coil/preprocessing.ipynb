```python
import pandas as pd
```

```python
df= pd.read_csv("2025-11-15_22-39-30-937.csv")
df
```

```python
df.CH3.plot()
```

```python
Ts= df.Ts.loc[0]
```

```python
df.CH4.plot()
```

```python
CH4 = df.CH4.values
CH4
```

```python
len(CH4)
```

```python
v = CH4-CH4[:300].mean()
```

```python
import numpy as np
```

```python
intV = np.cumsum(v)*Ts
df.CH3.plot()
(pd.Series(intV)*1.1e8).plot()
```
