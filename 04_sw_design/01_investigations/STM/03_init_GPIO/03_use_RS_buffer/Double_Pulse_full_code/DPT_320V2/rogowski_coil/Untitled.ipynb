```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("2025-11-16_12-24-58-791.csv")
df.head()
```

```python
Ts = df.Ts.values[0]
```

```python
vds = df.CH1.values*100 # x100 probe
il =  df.CH3.values*100 # x100 probe
il = il -il[:500].mean()
# rogowski coil
id_d= df.CH2.values
id_d = id_d -id_d[:500].mean()
Id = np.cumsum(id_d)*Ts
plt.plot(Id)
```

```python
import scipy 
Id= scipy.integrate.cumulative_trapezoid(y= id_d, dx =Ts, initial = 0)
plt.plot(Id)
```
