```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("df.csv")
df.plot(x="t",y="v")

s= df.v.rolling(window=10).mean()
plt.plot( df.t, s)
```

```python
df=pd.read_csv("df.csv")
df.plot(x="t",y="v")

s= df.v.rolling(window=100).median()
plt.plot( df.t, s)
```

```python
ax = plt.subplot()

df=pd.read_csv("df.csv")


df.loc[:,"v2"]= df.v.rolling(window=10).mean()
df.loc[:,"v3"]= df.v.rolling(window=10).median()

n=len(df)
df2=df.iloc[int(n*6/10):]

df2.plot(x="t",y="v", ax= ax,label="raw")
df2.plot(x="t",y="v2", ax= ax,label="mean W=10")
df2.plot(x="t",y="v3", ax= ax,label="median W=10")
```

```python
s= df2.v
plt.plot(s,label = "raw")
for w in [5,10,50,100,200]:
    s= s.rolling(window=w).mean()
    plt.plot(s, label = "W = "+str(w))
plt.legend()
```
