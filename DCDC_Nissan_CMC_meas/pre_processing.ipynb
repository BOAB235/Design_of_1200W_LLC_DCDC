```python
import pandas as pd
```

```python
df = pd.read_csv("L_meas.csv")
df.plot(x= "time", y=["CH2", "CH3"])
```

```python
t= df.time-df.time.values[0]
CH2= t.astype("str") + "\t"+ df.CH2.astype("str")
CH3= t.astype("str") + "\t"+ df.CH3.astype("str")
with open("CH2.txt", "w") as f:
    f.write("\n".join(CH2.values))
with open("CH3.txt", "w") as f:
    f.write("\n".join(CH3.values))
```
