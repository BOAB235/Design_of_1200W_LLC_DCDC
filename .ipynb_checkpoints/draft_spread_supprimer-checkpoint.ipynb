```python
arr=[]
n=1000
T=10e-6
k=0

for i in range(n//2): 
    arr.append(T+k*0.1*T/100)
    arr.append(T-k*0.1*T/100)
    k=k+1
    k=k%100
time = []
sig=[]
for dt in arr:
    time .append(dt/2)
    time .append(dt)
    sig.append(1)
    sig.append(-1)
time = np.array(time)
t = time.cumsum()
sig= np.array(sig)
plt.plot(t[:10], sig[:10])


```

```python
import numpy as np

t = t    # non-uniform time
x = sig     # signal

# oversampling factor
OS = 100

# uniform grid
N = len(t)
t_u = np.linspace(t.min(), t.max(), OS * N)

# interpolate signal
x_u = np.interp(t_u, t, x)

# sampling period
Ts = t_u[1] - t_u[0]

# FFT
Xu = np.fft.fft(x_u)
freq = np.fft.fftfreq(len(x_u), Ts)

# keep positive frequencies
idx = freq >= 0
freq = freq[idx]
mag = np.abs(Xu[idx]) / len(x_u)
plt.plot(freq,mag)
plt.plot([1/T,1/T],[1e-1,1e-7])
plt.xscale("log")
plt.yscale("log")
plt.grid()
```
