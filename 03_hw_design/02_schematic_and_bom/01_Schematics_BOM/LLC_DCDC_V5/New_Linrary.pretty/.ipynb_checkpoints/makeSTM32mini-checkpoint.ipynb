```python
txt_rec="""
	(fp_rect
		(start xs_ ys_)
		(end   xe_ ye_)
		(stroke
			(width 0.1)
			(type default)
		)
		(fill no)
		(layer "F.SilkS")
	)
"""

def make_rec(xs,ys,xe,ye):
    txtin = copy(txt_rec)
    txtin = txtin.replace("xs_",str(xs))
    txtin = txtin.replace("ys_",str(ys))
    txtin = txtin.replace("xe_",str(xe))
    txtin = txtin.replace("ye_",str(ye))
    return txtin
```

```python
import numpy as np

step = 2.54
txt_  = copy(txt)
x0, y0 = -10,10
L = 20.32
N= 10
H = (N-1)*step
xarr = np.array([0,step, L-step, L])+x0
for j,x in enumerate (xarr):
    for i in range(N):
        y= step*i+y0
        txt_+=make_pad((i+1)+10*j, x, y)
txt_+=make_rec(x0-step/2,y0-step/2,x0+L+step/2,y0+H+step/2)
txt_+="\n)"
with open("SMT32F1-MINI.kicad_mod","w") as f:
    f.write(txt_)
```
