# COLOR AND PLOT 

```python
import cadquery as cq

base   = cq.Workplane("XY").box(20, 20, 5)
pillar = cq.Workplane("XY").workplane(offset=5).circle(6).extrude(20)

assy = cq.Assembly(name="colored_parts")
assy.add(base,   name="base",   color=cq.Color(0.2, 0.6, 1.0))  # blue-ish
assy.add(pillar, name="pillar", color=cq.Color("green"))        # named color

assy.save("colored_assembly.step")  # colors preserved
```

```python
cube = cq.Workplane("XY").box(10, 10, 10)
sphere = cq.Workplane("XY").workplane(offset=8).sphere(5)

assy = cq.Assembly()
assy.add(cube,   name="cube",   color=cq.Color(1, 0, 0))  # red
assy.add(sphere, name="sphere", color=cq.Color(0, 1, 0))  # green

assy.save("colored_assembly.glb")
```
