```python
%%render 
A = 21.8
B =16.8
C=15.8
E=5
N= 4
xa=(A-B)/2
xb=A/2-E/2
xc=A/2+E/2
xd= A-xa
xab= abs(xa-xb)
xdc=xab
xcb=E
```

```python
margin = 1.3
eCu=0.5
dx = (xb-margin)-(xa+margin)
lineW= (dx-(N-1)*eCu)/N
lineW
```

```python
step = lineW+eCu
```

```python
xstart= xa+margin+lineW/2
ystart= 0+margin+lineW/2
Lx= xd-xa-2*margin-2*lineW/2
Ly= C+2*(N*lineW+eCu*(N-1)+margin)
```

```python
from copy import copy 
def trans(p,v):
    return p+v

def plt_line(p1,p2,width=1):
    x= [p1[0],p2[0]]
    y= [p1[1],p2[1]]
    plt.plot(x, y, 'b',linewidth=width)
```

```python
def plot_spiral (p1=np.array([0,0]),Lx=10,Ly=12,step = 1,N= 4,width=1,plot= True):
    #width = width_correction(width)

    points=[]


    vt=np.array([Lx,0])
    for i in range(4*N+1):
    
    
        
        #print(vt)
        p2= trans(p1,vt)
        if plot: 
            plt_line(p1,p2,width=width)
        points.append([p1,p2])
        p1 = copy(p2)
        
    
    
        
    
        match i%4 :
            case 0:
                vt=np.array([0,Ly])
                Ly=Ly-step
            case 1:
                vt=np.array([-Lx,0])
                Lx=Lx-step
            case 2:
                vt=np.array([0,-Ly])
                Ly=Ly-step
            case 3:
                vt=np.array([Lx,0])
                Lx=Lx-step
    return np.array(points).round(3).tolist()
    

    
```

```python
#plot_spiral (p1=np.array([0,0]),Lx=10,Ly=12,step = 1,N= 4,width=1)
points= plot_spiral (p1=np.array([0,0]),Lx=10,Ly=12,step = 1,N= 4,width=1)
plt.gca().set_aspect('equal')
#plt.legend()
plt.grid()
    
```
