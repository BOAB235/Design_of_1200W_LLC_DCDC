```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid #cumtrapz
from scipy.integrate import cumulative_trapezoid
```

```python
df= pd.read_csv("test_double_pulse_Draft_V3.txt", sep="\t")
t= df.time.values
vds= df["V(n003)"].values
ig= df["I(R1)"].values
plt.plot(t,vds)
plt.plot(t,ig*100)
```

```python
q = cumulative_trapezoid (y= ig, x= t,initial=0)
plt.plot(t,q)
```

```python
q2= q+0.8e-8*t/1e-5
plt.plot(t,q2)
```

```python
m1 = t>0.19e-5
m2= t<0.38e-5
t3= t[m1 & m2]
vds3= vds[m1 & m2]
ig3 = ig[m1 & m2]
q3 = q2[m1 & m2]
plt.plot(t3,q3)
```
