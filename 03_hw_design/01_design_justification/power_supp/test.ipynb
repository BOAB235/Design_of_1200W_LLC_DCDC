I_{\text{out}} = \frac{I_{\text{pk,sec}}}{2} \cdot (1 - D)
$$

Where:
- \( I_{\text{out}} \) = average DC output current after the capacitor
- \( I_{\text{pk,sec}} \) = peak secondary current
- \( D \) = duty cycle (fraction of ON time)

$$
I_{\text{pk,sec}} = I_{\text{pk,pri}} \cdot \frac{N_p}{N_s}

$$

Where:
- \( N_p, N_s \) = primary and secondary turns
- \( V_{\text{in}} \) = input voltage
- \( L_m \) = magnetizing inductance
- \( f \) = switching frequency

You’re on the right track with your intuition. Let’s derive the formula for **duty cycle ( D )** when we include the secondary copper resistance ( R\_s ) and diode drop ( V\_d ).

***

### ✅ Steady-State Flyback Voltage Relation

During OFF time, the secondary sees:
$$
V\_{\text{sec}} = V\_{\text{out}} + V\_d + I\_{\text{sec}} \cdot R\_s
$$

Reflected to the primary:
$$
V\_{\text{pri,ref}} = V\_{\text{sec}} \cdot \frac{N\_p}{N\_s}
$$

During ON time, the primary sees ( V\_{\text{in}} ). In steady state:
$$
V\_{\text{in}} \cdot D = V\_{\text{pri,ref}} \cdot (1-D)
$$

So:
$$
\frac{D}{1-D} = \frac{V\_{\text{pri,ref}}}{V\_{\text{in}}} = \frac{(V\_{\text{out}} + V\_d + I\_{\text{sec}} \cdot R\_s) \cdot \dfrac{N\_p}{N\_s}}{V\_{\text{in}}}
$$

***

### ✅ Solve for ( D ):

$$
D = \frac{\dfrac{(V\_{\text{out}} + V\_d + I\_{\text{sec}} \cdot R\_s) \cdot \dfrac{N\_p}{N\_s}}{V\_{\text{in}}}}{1 + \dfrac{(V\_{\text{out}} + V\_d + I\_{\text{sec}} \cdot R\_s) \cdot \dfrac{N\_p}{N\_s}}{V\_{\text{in}}}}
$$

# ELP 22/6/16 with I 22/2.5/16 N87 with air gap

```python
import pandas as pd
import numpy as np

# --- Input Parameters ---
Vin = 400           # Input Voltage (V)
Vout = 12           # Output Voltage (V)
Vd = 0.7            # Diode Forward Drop (V)
f = 132e3           # Switching Frequency (Hz)
Bs = 0.3            # Saturation Flux Density (T)
Ae = 78.5e-6        # Effective Cross Section Area (m^2)
le = 26.1e-3        # Magnetic Path Length (m) for ELP 22/6/16
ui = 1360           # Initial Permeability (N87 material)
u0 = 4 * np.pi * 1e-7 # Permeability of Free Space


# mm2=>m2 ELP 22/6/16 with I 22/2.5/16 N87
# https://www.tdk-electronics.tdk.com/inf/80/db/fer/elp_22_6_16.pdf

# Target AL for a gapped core (nH/N^2)
# We choose a lower AL to prevent saturation
#AL_target = 250 


# --- 1. Add these variables at the top ---
lg_target = 0.0002  # 0.2 mm in meters
# Calculate AL based on the physical gap
AL_target = (u0 * Ae / (lg_target + (le / ui))) * 1e9  # Result in nH

T = 1 / f
arr = []

for Np in range(20, 51, 2):  # Primary turns
    for Ns in [3, 4, 5, 8]:     # Secondary turns
        # 1. Recalculate Duty Cycle (D) based on Turns Ratio
        n = Np / Ns
        Vro = n * (Vout + Vd)
        D = Vro / (Vin + Vro)
        
        if D > 0.5: continue # Standard Flyback limit
        
        Ton = D * T
        
        # 2. Correct Bpeak calculation (No division by 2 for Flyback)
        Bpeak = (Vin * Ton) / (Np * Ae)
        
        # 3. Magnetizing Inductance (Lm)
        Lm = AL_target * (Np**2) * 1e-9 # convert nH to Henrys
        
        # 4. Peak Currents
        Ippeak = (Vin * Ton) / Lm
        Ispeak = Ippeak * n
        
        # 5. Air Gap Calculation (lg)
        # Formula: lg = (u0 * Np^2 * Ae / Lm) - (le / ui)
        lg = ((u0 * (Np**2) * Ae) / Lm) - (le / ui)
        
        # 6. Output Power (DCM assumption)
        Effec = 0.8
        Ps = 0.5 * Lm * (Ippeak**2) * f * Effec
        
        Margin = (Bs - Bpeak) / Bs
        
        if Margin >= 0.15: # Only keep designs with 15% flux margin
            arr.append({
                "Np": Np, "Ns": Ns, "D": round(D, 3), 
                "Bpeak": round(Bpeak, 3), "Lm_uH": round(Lm*1e6, 1),
                "Ip_peak": round(Ippeak, 2), "Ps_W": round(Ps, 1),
                "Gap_mm": round(lg * 1000, 3) , # convert to mm
                "Margin": Margin
            })

df = pd.DataFrame(arr).sort_values('Bpeak')
print(df.to_string(index=False))
```
