```python
def trigger (sig , thresh= None, transition= "rise", marginI = 20):
    arr=[0]
    if thresh is None: 
        thresh = (sig.max()+sig.min())/2
    for i in range(1, len(sig)):
        if transition =="rise":
            if sig[i-1] <thresh and sig[i] >= thresh and i > arr[-1]+marginI:
                arr.append(i)
        elif sig[i-1] >thresh and sig[i] <= thresh and i > arr[-1]+marginI:
            arr.append(i)
    return arr[1:]
```

## Capacitor discharging

### Setup

The setup is relatively simple: a power supply (400 V) is connected to the HV connector, and a relay is used to open the circuit and measure the discharge voltage.

```python
fig_counter=One_figure_with_captions(fig_counter, img1="Setup1.png"
                                     , caption1="Capacitor discharge setup",
                                     first_part_path="figs/", 
                                     width=19)
```

In this DC‑DC converter, the flyback UVLO resistor is removed, so the flyback remains active down to 30 V and therefore acts as an active discharge path. 

For safety, the DC input voltage measurement divider is designed to discharge the input even if the flyback fails.

### Test of normal operation

In this test, the Rload is 27 $\ohm$ (around 85W)

### Passive discharge, only discharge resistors

Because the auxiliary flyback cannot be disabled by design, this final case will be evaluated only through simulation. The capacitor and resistor values are selected to model the worst‑case scenario.

```python
fig_counter=One_figure_with_captions(fig_counter, img1="sim.png"
                                     , caption1="Capacitor discharge setup",
                                     first_part_path="figs/", 
                                     width=19)
```

In this simulation, we considered the worst case for all components:
- DC-Link capacitor: 10% according to the datashet.
- Resonant capacitor: 5% according to the datashet.
- EMI filter differential mode capacitor: 10% according to the datashet.
- Discharge resistor (the same resistor used in the measurement divider): 1%, while the datasheet specifies only 0.1%

The discharge time between 400 V and 60 V is 3.6 s, fully satisfying the 5‑second requirement.

Even with 600V, the discharge time is 4.4s => satisfying the 5‑second requirement (see below)
