* TinySwitch-II is an offline **flyback switcher with an internal MOSFET**, not a classic PWM controller.
* It works in **current-mode**: each cycle starts with MOSFET ON, current ramps up, then turns OFF at current limit or DCMAX.
* The internal oscillator runs at **~138 kHz**, but the **effective switching frequency is variable**.
* Output regulation is done by **cycle skipping**, not by duty-cycle control.
* The **EN/UV pin is current-controlled**, not voltage-controlled; >240 A drawn disables a cycle.
* The optocoupler **steals current from EN/UV** to command skipping or current-limit reduction.
* At light load, cycles are skipped to save power and avoid audible noise.
* At heavy load, cycles are **not skipped**; instead the **current limit is reduced** by an internal state machine.
* Maximum output power is set mainly by **transformer primary inductance and peak current**.
* The **BP capacitor stabilizes the internal reference**, indirectly allowing full current limit and rated power.
