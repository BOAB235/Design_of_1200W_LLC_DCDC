```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
df = pd.read_csv('2026-01-25_12-42-25-298.csv')
Ts= df.Ts.iloc[0]
Vds= df.CH1.values
Vcr = df.CH4.values
time = np.arange(0,len(Vds)*Ts, Ts)
```

```python
plt.plot(time, Vcr)
plt.plot(time, Vds)
```

```python
M= Vcr.max()
m= Vcr.min()
Vcr2= Vcr-(M+m)/2
plt.plot(time, Vcr2)
```

```python
# i = c dv /dt
Cr=120e-9
ic = np.cumsum(Vcr2)*Ts/Cr
ic2= ic-(ic.max()+ic.min())/2
plt.plot(ic*1e-1+30)
plt.plot(Vds)
plt.plot(Vcr2)
```
